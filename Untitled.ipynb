```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
RADIUS=21
```

```python
im1 = r"/mnt/c/Users/James/test_camera_pics/img2021-07-04 03-23-12.439031.jpg"
im2 = r"/mnt/c/Users/James/test_camera_pics/img2021-07-04 03-24-12.979465.jpg"
```

```python
image1 = cv2.imread(im1)
orig1 = image1.copy()
gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
image2 = cv2.imread(im2)
orig2 = image2.copy()
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
```

```python
T# apply a Gaussian blur to the image then find the brightest
# region
gray1 = cv2.GaussianBlur(gray1, (RADIUS, RADIUS), 0)
(minVal1, maxVal1, minLoc1, maxLoc1) = cv2.minMaxLoc(gray1)
cv2.circle(image1, maxLoc1, RADIUS, (255, 0, 0), 2)
plt.imshow(image1)
plt.show()
```

```python
gray2 = cv2.GaussianBlur(gray2, (RADIUS, RADIUS), 0)
(minVal2, maxVal2, minLoc2, maxLoc2) = cv2.minMaxLoc(gray2)
cv2.circle(image2, maxLoc2, RADIUS, (255, 0, 0), 2)
plt.imshow(image2)
plt.show()
```
